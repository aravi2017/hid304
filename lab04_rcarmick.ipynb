```python
length_perceived = []
length_actual = []
```

```python
a = random.uniform(0.01, 0.1)
%matplotlib inline
plt.barh(0.25, 0.01, 0.5, align = 'center', color = 'blue')
plt.barh(1.25, a, 0.5, align = 'center', color = 'blue')
plt.barh(2.25, 0.1, 0.5, align = 'center',color = 'blue')
plt.gca().set_frame_on(True)
plt.gca().get_xaxis().set_visible(False)
plt.yticks(np.arange(3)+0.25, ('1', '?', '10'))
```

```python
guess = eval(input("What is your guess for the length: "))
length_perceived.append(guess)
length_actual.append(a)
```

```python
print(length_perceived)
print(length_actual)
```

```python
plt.scatter(length_perceived, length_actual)
plt.title("Is Steven's Power Law true? (10)")
plt.xlabel("Actual")
plt.ylabel("Perceived")
```

```python
from scipy.optimize import curve_fit

def func(x, a, b, c):
    return(a * np.power(x, b) + c)
x = length_perceived
y = length_actual
popt, pcov = curve_fit(func, x, y)
print('{:.2f} x^{:.2f} + {:.2f}'.format(*popt))
```

```python
size_perceived = []
size_actual = []
```

```python
import math as math

b = random.uniform(0.01, 0.1)
radius1 = math.sqrt(0.01/math.pi)
radius2 = math.sqrt(b/math.pi)
radius3 = math.sqrt(0.1/math.pi)

plt.axis('equal')
plt.axis('off')

circ1 = plt.Circle( (0,0),         radius1, clip_on=False )
circ2 = plt.Circle( (2*radius2,0), radius2, clip_on=False )
circ3 = plt.Circle( (4*radius3,0), radius3, clip_on=False )

plt.gca().add_artist(circ1)
plt.gca().add_artist(circ2)
plt.gca().add_artist(circ3)
```

```python
size_guess = eval(input("What is the size of the unknown circle?: "))
size_perceived.append(size_guess)
size_actual.append(b)
```

```python
print(size_perceived)
print(size_actual)
```

```python
plt.scatter(size_perceived, size_actual)
plt.title("Is Steven's Power Law true? (10)")
plt.xlabel("Actual")
plt.ylabel("Perceived")
```

```python
def func(x, a, b, c):
    return(a * np.power(x, b) + c)
x = size_perceived
y = size_actual
popt, pcov = curve_fit(func, x, y)
print('{:.2f} x^{:.2f} + {:.2f}'.format(*popt))
```
